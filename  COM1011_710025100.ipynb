## 14.  If you run the code from questions 10, 11, and 12 again, do the scores (precision, recall, accuracy, F1) change? Why (or why not)? Please explain where those scores come from, and how do they change (or don't change) if you run the code again. 
[3 marks]

The scores do change each time the code is run. This is because each time the code is ran, the training test and the test sets are allocated differently. 

The precision score is calculated by true positives/(true positives + false positives). The precision score shows the ability of the classifier to not label a sample which is negative as positive.The best value is 1 and the worst value is 0.

The Recall score is calculated by true positives/(true positives + false negatives). The Recall score shows the ability of the classifier to find all the positive samples. The best value is 1 and the worst value is 0.

The accuracy score is calculated by number of correct predictions/total number of predictions made. The best value is 1 and the worst value is 0.

The f1 score is the weighted average of precision and recall. It is calculated by 2*(precision * recall)/(precision + recall). The best value is 1 and the worst value is 0.  


when the code is re-run the scores either slightly increase or decrease. This is because the split function seperates the training and test set randomly according to a random sequence.

## 31. Usually, DBSCAN takes longer than K-means to run, and the time it takes to run is affected by the eps parameter. Explain why that is the case.
[4 marks]

DBSCAN takes longer than K-means to run because DBSCAN needs an EPS value. An EPS value is the lowest distance a point can be away from a cluster in order for it to still be counted as part of the cluster. This means that having a high value of EPS for example, will make the run time shorter than if the EPS was a small value (and vice versa). This is because a high EPS value will take less CPU cycles to compute. Furthermore, Kmeans only requires the number of clusters to be inputted and therefore requires less trial and error for the computer to find the best fit.
